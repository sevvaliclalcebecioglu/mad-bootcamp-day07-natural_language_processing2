```python
cumleler= [['Hello','This','is','python','training','by','Aman'],

             ['Hello','This','is','Java','training','by','Aman'],

             ['Hello','This','is','Data Science','training','by','Unfold','Data','Science'],

             ['Hello','This','is','programming','training','']]

# Word embedding ve NLP √ßalƒ±≈ümalarƒ±nda kullanƒ±lmak √ºzere tokenlara ayrƒ±lmƒ±≈ü √∂rnek c√ºmleleri i√ßeren bir liste olu≈üturur
```

```python
# LDA = Latent Drichlet Allocation Topic Modelling / C√ºmlenin hangi konuda konustuƒüunu anlƒ±yor

# LDA, metin dok√ºmanlarƒ± i√ßerisindeki gizli konularƒ± (topic) olasƒ±lƒ±ksal olarak ke≈üfetmek i√ßin kullanƒ±lan bir konu modelleme y√∂ntemidir.
```

```python
import gensim

# Gensim k√ºt√ºphanesi, konu modelleme ve metin madenciliƒüi i≈ülemleri i√ßin kullanƒ±lƒ±r


from gensim import corpora

# Metinleri sayƒ±sal s√∂zl√ºk (dictionary) yapƒ±sƒ±na d√∂n√º≈üt√ºrmek i√ßin kullanƒ±lƒ±r


from gensim.models import LdaModel

# LDA konu modelini olu≈üturmak i√ßin gerekli sƒ±nƒ±f i√ße aktarƒ±lƒ±r



# Sample documents

documents = [

    "Tennis is my favorite sport to play.",

    "Football is a popular competition in certain country.",

    "There are many athletes currently training for the olympic."

]

# Konu modelleme yapƒ±lacak √∂rnek metin dok√ºmanlarƒ± tanƒ±mlanƒ±r



# Preprocess documents

texts = [[word for word in document.lower().split()] for document in documents]

# Metinler k√º√ß√ºk harfe √ßevrilir ve kelime bazlƒ± listelere ayrƒ±lƒ±r


dictionary = corpora.Dictionary(texts)

# Kelimelerden benzersiz bir s√∂zl√ºk (id‚Äìkelime e≈ülemesi) olu≈üturulur


corpus = [dictionary.doc2bow(text) for text in texts]

# Her dok√ºman, kelime‚Äìfrekans √ßiftlerinden olu≈üan sayƒ±sal formata d√∂n√º≈üt√ºr√ºl√ºr



# The LDA model

lda_model = LdaModel(corpus, num_topics=2, id2word=dictionary, passes=15)

# 2 konu (topic) olacak ≈üekilde LDA modeli eƒüitilir


topics = lda_model.print_topics()

# Modelin √∂ƒürendiƒüi konular ve bu konulara ait kelimeler alƒ±nƒ±r


for topic in topics:

    print(topic)

# Her bir konunun i√ßeriƒüi ekrana yazdƒ±rƒ±lƒ±r
```

```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import gensim
from gensim import corpora
from gensim.models import LdaModel

# Gerekli web scraping ve topic modeling k√ºt√ºphaneleri i√ße aktarƒ±lƒ±r


# Function to scrape transcript data
def url_to_transcript(url):
    '''Returns transcript text specifically from scrapsfromtheloft.com.'''

    # Verilen URL'den HTML i√ßeriƒüini √ßeker
    page = requests.get(url).text

    # HTML i√ßeriƒüi BeautifulSoup ile parse edilir
    soup = BeautifulSoup(page, "lxml")

    # Sayfa i√ßerisindeki konu≈üma metinleri paragraf bazƒ±nda alƒ±nƒ±r
    text = [p.text for p in soup.find(class_="ast-container").find_all('p')]

    # Ba≈üarƒ±lƒ± scraping i≈ülemi kullanƒ±cƒ±ya bildirilir
    print(f'Successfully scraped: {url}')

    # Paragraflar birle≈ütirilerek tek bir metin olarak d√∂nd√ºr√ºl√ºr
    return ' '.join(text)


# URLs of transcripts
urls = [
    'http://scrapsfromtheloft.com/2017/05/06/louis-ck-oh-my-god-full-transcript/',
    'http://scrapsfromtheloft.com/2017/04/11/dave-chappelle-age-spin-2017-full-transcript/',
    'http://scrapsfromtheloft.com/2018/03/15/ricky-gervais-humanity-transcript/',
    'https://scrapsfromtheloft.com/comedy/bo-burnham-what-transcript/',
    'http://scrapsfromtheloft.com/2017/05/24/bill-burr-im-sorry-feel-way-2014-full-transcript/',
    'http://scrapsfromtheloft.com/2017/04/21/jim-jefferies-bare-2014-full-transcript/',
    'http://scrapsfromtheloft.com/2017/08/02/john-mulaney-comeback-kid-2015-full-transcript/',
    'https://scrapsfromtheloft.com/comedy/hasan-minhaj-homecoming-king-transcript/',
    'http://scrapsfromtheloft.com/2017/09/19/ali-wong-baby-cobra-2016-full-transcript/',
    'http://scrapsfromtheloft.com/2017/08/03/anthony-jeselnik-thoughts-prayers-2015-full-transcript/',
    'http://scrapsfromtheloft.com/2018/03/03/mike-birbiglia-my-girlfriends-boyfriend-2013-full-transcript/',
    'http://scrapsfromtheloft.com/2017/08/19/joe-rogan-triggered-2016-full-transcript/'
]

# Stand-up komedyenlerine ait konu≈üma metinlerinin URL listesi tanƒ±mlanƒ±r


# Comedian names (for reference)
comedians = ['louis', 'dave', 'ricky', 'bo', 'bill', 'jim', 'john', 'hasan', 'ali', 'anthony', 'mike', 'joe']

# Metinlerin hangi komedyene ait olduƒüunu temsil eden referans listesi olu≈üturulur


# Scrape transcripts
documents = [url_to_transcript(url) for url in urls]

# Her URL i√ßin konu≈üma metinleri √ßekilerek documents listesine aktarƒ±lƒ±r


# Preprocess the transcripts for LDA modeling
texts = [
    gensim.utils.simple_preprocess(doc, deacc=True)
    for doc in documents
]

# Metinler k√º√ß√ºk harfe √ßevrilir, noktalama i≈üaretleri temizlenir ve tokenize edilir


# Create dictionary and corpus
dictionary = corpora.Dictionary(texts)

# Metinlerde ge√ßen kelimelerden s√∂zl√ºk (dictionary) olu≈üturulur

corpus = [dictionary.doc2bow(text) for text in texts]

# Her dok√ºman kelime-frekans (Bag of Words) formatƒ±na d√∂n√º≈üt√ºr√ºl√ºr


# LDA Model setup
lda_model = LdaModel(
    corpus,
    num_topics=5,
    id2word=dictionary,
    passes=15,
    random_state=42
)

# Konu≈üma metinlerinden 5 ana konu √ßƒ±karmak i√ßin LDA modeli eƒüitilir


# Print the topics
topics = lda_model.print_topics(num_words=10)

# Her konuya ait en √∂nemli 10 kelime alƒ±nƒ±r

print("\nTopics discovered in comedians' transcripts:")

# LDA modelinin bulduƒüu konular ekrana yazdƒ±rƒ±lƒ±r

for idx, topic in topics:
    print(f"Topic #{idx + 1}: {topic}\n")

# Her bir konunun i√ßeriƒüi sƒ±rasƒ±yla kullanƒ±cƒ±ya g√∂sterilir
```

```python
# Imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
import gensim
from gensim import corpora
from gensim.models.ldamulticore import LdaMulticore
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import text
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Scrape function
def url_to_transcript(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    paragraphs = soup.find(class_="ast-container").find_all('p')
    text_data = ' '.join(p.text for p in paragraphs)
    print(f'Successfully scraped: {url}')
    return text_data

# Custom stopwords
add_stop_words = [
    'like', 'im', 'know', 'just', 'dont', 'thats', 'right', 'people', 'youre', 'got',
    'gonna', 'time', 'think', 'yeah', 'said', 'want', 'say', 'thing', 'going', 'shit',
    'fucking', 'didnt', 'hes', 'shes', 'theyre', 'weve', 'ill', 'youve', 'ive', 'couldnt',
    'wouldnt', 'cant', 'wont', 'one', 'get', 'really', 'see', '‚ô™', 'g', 'okay'
]

# Convert frozenset to list for CountVectorizer
stop_words = list(text.ENGLISH_STOP_WORDS.union(add_stop_words))


# Lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to extract only nouns and adjectives
def nouns_adj(text):
    text = text.lower()
    text = re.sub(r"‚Äô", "'", text)
    text = re.sub(r"[^a-zA-Z']", " ", text)
    text = re.sub(r'\s+', ' ', text)
    tokenized = word_tokenize(text)
    tagged = pos_tag(tokenized)
    is_noun_adj = lambda pos: pos.startswith('NN') or pos.startswith('JJ')
    words = [lemmatizer.lemmatize(word) for word, pos in tagged if is_noun_adj(pos) and word not in stop_words]
    return ' '.join(words)

# URLs and comedian names
urls = [
    'http://scrapsfromtheloft.com/2017/05/06/louis-ck-oh-my-god-full-transcript/',
    'http://scrapsfromtheloft.com/2017/04/11/dave-chappelle-age-spin-2017-full-transcript/',
    'http://scrapsfromtheloft.com/2018/03/15/ricky-gervais-humanity-transcript/',
    'https://scrapsfromtheloft.com/comedy/bo-burnham-what-transcript/',
    'http://scrapsfromtheloft.com/2017/05/24/bill-burr-im-sorry-feel-way-2014-full-transcript/',
    'http://scrapsfromtheloft.com/2017/04/21/jim-jefferies-bare-2014-full-transcript/',
    'http://scrapsfromtheloft.com/2017/08/02/john-mulaney-comeback-kid-2015-full-transcript/',
    'https://scrapsfromtheloft.com/comedy/hasan-minhaj-homecoming-king-transcript/',
    'http://scrapsfromtheloft.com/2017/09/19/ali-wong-baby-cobra-2016-full-transcript/',
    'http://scrapsfromtheloft.com/2017/08/03/anthony-jeselnik-thoughts-prayers-2015-full-transcript/',
    'http://scrapsfromtheloft.com/2018/03/03/mike-birbiglia-my-girlfriends-boyfriend-2013-full-transcript/',
    'http://scrapsfromtheloft.com/2017/08/19/joe-rogan-triggered-2016-full-transcript/'
]

comedians = ['Louis CK', 'Dave Chappelle', 'Ricky Gervais', 'Bo Burnham', 'Bill Burr',
             'Jim Jefferies', 'John Mulaney', 'Hasan Minhaj', 'Ali Wong',
             'Anthony Jeselnik', 'Mike Birbiglia', 'Joe Rogan']

# Scrape and clean
documents = [url_to_transcript(url) for url in urls]
cleaned_docs = [nouns_adj(doc) for doc in documents]

# Build CountVectorizer
cv = CountVectorizer(stop_words=stop_words, max_df=0.8)
data_cv = cv.fit_transform(cleaned_docs)

# Convert into pandas DataFrame for gensim
dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names_out())

# Create corpus for LDA
corpus = gensim.matutils.Sparse2Corpus(data_cv, documents_columns=False)
id2word = dict((id, word) for word, id in cv.vocabulary_.items())

# LDA Model (12 topics for 12 comedians)
lda_model = LdaMulticore(corpus, num_topics=12, id2word=id2word, passes=50, workers=2, random_state=42)

# Show topics
topics = lda_model.print_topics(num_words=10)
for idx, topic in topics:
    print(f"\nüìù Topic {idx}:\n{topic}")
```

```python
doc=nlp('Jonathan is looking for a job in Turkey for 50k. World Health Organization WHO')

# Verilen metni NLP pipeline‚Äôdan ge√ßirerek token, entity ve dilsel √∂zellikleri analiz eder
```
